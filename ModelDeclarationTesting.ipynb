# Part 1

Function for plotting the datasets

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_2d_dataset(X, y, title="2D Dataset", xlabel="Feature 1", ylabel="Feature 2"):
    plt.figure(figsize=(8, 6))
    scatter = plt.scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", edgecolors="k", s=60)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.legend(*scatter.legend_elements(), title="Classes")
    plt.show()
```

Generating a bidimensional dataset

```python
from sklearn.datasets import make_classification


# Generate bidimensional dataset
X, y = make_classification(
    n_samples=200,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_classes=2,
    random_state=0
)


```

plotting the data set

```python
plot_2d_dataset(X, y, title="Make Classification 2D Data", xlabel="X-axis", ylabel="Y-axis")
```

# Model Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier



# Train logistic regression
logRegModel = LogisticRegression()

# Train Linear Discriminant Analysis
ldaModel = LinearDiscriminantAnalysis()

# Train Quadratic Discriminant Analysis
qdaModel = QuadraticDiscriminantAnalysis()

# Train SVM with linear kernel
swmLinModel = SVC(kernel='linear')

# Train SVM with RBF kernel
swmRBFModel = SVC(kernel='rbf')

# Train Decision Tree without pruning
dtNoPrune = DecisionTreeClassifier()

# Train Decision Tree with a maximum depth of 2
dtDepth2 = DecisionTreeClassifier(max_depth=2)
```

# Model Testing with Hold Out

```python
from sklearn.model_selection import train_test_split

# Split the data into train and test
X_train, X_test, y_train, y_test = train_test_split(
   X, y, test_size=0.3, random_state=0
)

logRegModel.fit(X_train, y_train)
ldaModel.fit(X_train, y_train)
qdaModel.fit(X_train, y_train)
swmLinModel.fit(X_train, y_train)
swmRBFModel.fit(X_train, y_train)
dtNoPrune.fit(X_train, y_train)
dtDepth2.fit(X_train, y_train)
```

```python
from sklearn.metrics import f1_score


# Predict on test set
y_pred = logRegModel.predict(X_test)

# Evaluate using F1 score
f1 = f1_score(y_test, y_pred)
print(f"F1 Score: {f1:.4f}")


# Predict on test set
y_pred = ldaModel.predict(X_test)

# Evaluate using F1 score
f1 = f1_score(y_test, y_pred)
print(f"F1 Score: {f1:.4f}")


# Predict on test set
y_pred = qdaModel.predict(X_test)

# Evaluate using F1 score
f1 = f1_score(y_test, y_pred)
print(f"F1 Score: {f1:.4f}")

# Predict on test set
y_pred = swmLinModel.predict(X_test)

# Evaluate using F1 score
f1 = f1_score(y_test, y_pred)
print(f"F1 Score: {f1:.4f}")

# Predict on test set
y_pred = swmRBF.predict(X_test)

# Evaluate using F1 score
f1 = f1_score(y_test, y_pred)
print(f"F1 Score: {f1:.4f}")


# Predict on test set
y_pred = dtNoPrune.predict(X_test)

# Evaluate using F1 score
f1 = f1_score(y_test, y_pred)
print(f"F1 Score: {f1:.4f}")

# Predict on test set
y_pred = dtDepth2.predict(X_test)

# Evaluate using F1 score
f1 = f1_score(y_test, y_pred)
print(f"F1 Score: {f1:.4f}")
```

# Model Testing With Cross Validation

```python
from sklearn.model_selection import cross_val_score

f1_scores = cross_val_score(logRegModel, X, y, scoring='f1', cv=5)
print("Mean F1 score:", f1_scores.mean(), " - Logistic Regression")


f1_scores = cross_val_score(ldaModel, X, y, scoring='f1', cv=5)
print("Mean F1 score:", f1_scores.mean(), " - LDA")


f1_scores = cross_val_score(qdaModel, X, y, scoring='f1', cv=5)
print("Mean F1 score:", f1_scores.mean(), " - QDA")


f1_scores = cross_val_score(swmLinModel, X, y, scoring='f1', cv=5)
print("Mean F1 score:", f1_scores.mean(), " - SWM Linear")


f1_scores = cross_val_score(swmRBFModel, X, y, scoring='f1', cv=5)
print("Mean F1 score:", f1_scores.mean(), " - SWM RBF")


f1_scores = cross_val_score(dtNoPrune, X, y, scoring='f1', cv=5)
print("Mean F1 score:", f1_scores.mean(), " - Decision Trees no pruning")


f1_scores = cross_val_score(dtDepth2, X, y, scoring='f1', cv=5)
print("Mean F1 score:", f1_scores.mean(), " - Decision Trees Max Depth 2")
```
