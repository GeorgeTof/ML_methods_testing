## Synthetic 2D datasets from notebook examples

Bidimensional axis-aligned data 
- for the decision trees

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Generate a bidimensional dataset
np.random.seed(42)
X = np.random.rand(100, 2)
y = (X[:, 0] > 0.5).astype(int)  # Axis-aligned boundary at x = 0.5

# Train a decision tree model
model = DecisionTreeClassifier()
model.fit(X, y)

# Plot the dataset
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
plt.axvline(x=0.5, color='gray', linestyle='--', label='Boundary at x = 0.5')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('Axis-Aligned Dataset')
plt.show()

# Plot the decision tree
plt.figure(figsize=(12, 8))
plot_tree(model, feature_names=['Feature 1', 'Feature 2'], class_names=['Class 0', 'Class 1'], filled=True)
plt.show()
```

2D data with circular separation between classes
- for decision trees

```python
# Generate a bidimensional dataset
np.random.seed(42)
X = np.random.rand(100, 2)
# Circular decision boundary (center at (0.5, 0.5), radius 0.3)
y = ((X[:, 0] - 0.5) ** 2 + (X[:, 1] - 0.5) ** 2 < 0.3 ** 2).astype(int)

# Train a decision tree model
model = DecisionTreeClassifier()
model.fit(X, y)

# Plot the dataset
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
circle = plt.Circle((0.5, 0.5), 0.3, color='gray', fill=False, linestyle='--', label='Circular boundary')
plt.gca().add_patch(circle)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('Circular Decision Boundary')
plt.show()

# Plot the decision tree
plt.figure(figsize=(12, 8))
plot_tree(model, feature_names=['Feature 1', 'Feature 2'], class_names=['Class 0', 'Class 1'], filled=True)
plt.show()
```

```python
# check the results
preds = model.predict(X)
(vals,ans)=np.unique(preds==y, return_counts=True)
ans[vals]/sum(ans)
```
